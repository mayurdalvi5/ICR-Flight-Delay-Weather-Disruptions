```python
flights_df = pd.read_csv('/content/merged_Jan_2024_with_departure_datetime.csv')  # Weather CSV file path
weather_df = pd.read_csv('/content/weather_data_20240101_20240201_cleaned_15min.csv')  # Flights CSV file path
```

```python
# Convert `departure_resample15` to datetime format `YYYY-MM-DD HH:MM:SS`
flights_df['departure_resample15'] = pd.to_datetime(flights_df['departure_resample15'], format='%m/%d/%Y %H:%M')

# Convert `resample15` to datetime format `YYYY-MM-DD HH:MM:SS`
weather_df['resample15'] = pd.to_datetime(weather_df['resample15'])
```

```python
# Perform the left join on the entire flights_df, matching on 'rounded_departure', 'rounded_valid', and 'ORIGIN' with 'station'
merged_df = pd.merge(
    flights_df,
    weather_df,
    left_on=['departure_resample15', 'ORIGIN'],
    right_on=['resample15', 'station'],
    how='left'
)
```
